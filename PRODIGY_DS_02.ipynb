```python
# Rename our columns
df = df.rename(columns={
        'age':'Age',
        'sex':'Sex',
        'marital_status':'Marital_Status',
        'residence':'Residence',
        'occupation':'Occupation',
        'family_members':'Family_Members',
        'family_disabilities':'Family_Disabilities',
        'red_zone_frequency':'Red_Zone_Frequency',
        'n_vehicles':'No_Vehicles',
        'vehicle_type':'Vehicle_Type',
        'property_house':'Property_House'
})
```
